# Step 3: Exploratory Data Analysis (EDA)
## IBM Telco Customer Churn Dataset

**Objective:** Discover patterns, relationships, and business drivers of churn.

### Analyses included
- Overall churn distribution
- Churn rate by Contract, Payment Method, Internet Service
- Tenure distribution and binned churn rates
- Numeric box plots and correlation heatmap
- Demographic and billing insights

```python
import sys
from pathlib import Path

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.insert(0, str(PROJECT_ROOT / "src"))

from config import REPORTS_DIR, TARGET_COLUMN
from data_loader import load_cleaned_data, load_raw_data
from data_cleaning import clean_data, save_cleaned_data
from eda import ensure_churn_numeric, churn_rate_table, run_eda

# Load cleaned data (auto-clean if needed)
try:
    df = load_cleaned_data()
except FileNotFoundError:
    df_raw = load_raw_data()
    df, _ = clean_data(df_raw)
    save_cleaned_data(df)

df = ensure_churn_numeric(df)
print(f"Shape: {df.shape} | Churn rate: {df['ChurnNumeric'].mean()*100:.2f}%")
```

```python
# Churn rate by key features
for col in ["Contract", "PaymentMethod", "InternetService"]:
    print(f"\n--- {col} ---")
    display(churn_rate_table(df, col))
```

```python
# Generate and save all EDA plots
results = run_eda(df, REPORTS_DIR)
print("Saved figures:")
for p in results["figures"]:
    print(f"  {p.name}")
```
