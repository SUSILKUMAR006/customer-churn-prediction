# Step 2: Data Cleaning
## IBM Telco Customer Churn Dataset

**Objective:** Transform raw data into analysis-ready form.

### Cleaning actions
1. Strip whitespace from text fields
2. Fix data types
3. Remove duplicates
4. Impute missing `TotalCharges` (new customers → 0)
5. Validate categorical values
6. Check logical consistency
7. Document outliers (IQR — not removed)

```python
import sys
from pathlib import Path

import pandas as pd

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.insert(0, str(PROJECT_ROOT / "src"))

from data_loader import load_raw_data
from data_cleaning import clean_data, save_cleaned_data
from config import CLEANED_DATA_FILE

df_raw = load_raw_data()
print(f"Raw shape: {df_raw.shape}")
print(f"Missing TotalCharges: {df_raw['TotalCharges'].isna().sum()}")
```

```python
df_clean, report = clean_data(df_raw)

print("Cleaning report:")
for key, value in report.items():
    if key != "outlier_summary":
        print(f"  {key}: {value}")

print("\nOutlier summary:")
report["outlier_summary"]
```

```python
# Verify no missing values remain
assert df_clean.isnull().sum().sum() == 0, "Missing values detected after cleaning!"
print(f"Final shape: {df_clean.shape}")
df_clean.info()
```

```python
save_cleaned_data(df_clean)
print(f"Saved to: {CLEANED_DATA_FILE}")
df_clean.head()
```
