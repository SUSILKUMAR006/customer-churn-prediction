## IBM Telco Customer Churn Dataset

**Project:** Customer Churn Prediction (Hackathon)

### Why Churn Prediction Matters
- Retaining customers is **5–7× cheaper** than acquiring new ones in telecom.
- Churn directly impacts **Monthly Recurring Revenue (MRR)** and lifetime value.
- Proactive retention depends on **early risk detection**.

### Dataset Source
- **Kaggle:** [Telco Customer Churn](https://www.kaggle.com/datasets/blastchar/telco-customer-churn)
- **File:** `data/raw/Telco-Customer-Churn.csv` (7,043 customers × 21 columns)

## Feature Dictionary

| Feature | Type | Business Meaning |
|---------|------|------------------|
| customerID | ID | Unique identifier (excluded from modeling) |
| gender | Categorical | Male / Female |
| SeniorCitizen | Binary (0/1) | 1 if customer is 65+ |
| Partner | Categorical | Has partner or not |
| Dependents | Categorical | Has dependents or not |
| tenure | Numeric | Months with the company |
| PhoneService | Categorical | Phone service subscribed |
| MultipleLines | Categorical | Multiple phone lines |
| InternetService | Categorical | DSL / Fiber optic / No |
| OnlineSecurity | Categorical | Security add-on |
| OnlineBackup | Categorical | Backup add-on |
| DeviceProtection | Categorical | Device protection add-on |
| TechSupport | Categorical | Tech support add-on |
| StreamingTV | Categorical | Streaming TV add-on |
| StreamingMovies | Categorical | Streaming movies add-on |
| Contract | Categorical | Month-to-month / One year / Two year |
| PaperlessBilling | Categorical | Paperless billing flag |
| PaymentMethod | Categorical | Payment channel |
| MonthlyCharges | Numeric | Monthly bill amount ($) |
| TotalCharges | Numeric | Lifetime charges ($) |
| **Churn** | **Target** | **Yes = left, No = stayed** |

```python
# Target variable distribution
churn_counts = df[TARGET_COLUMN].value_counts()
churn_pct = df[TARGET_COLUMN].value_counts(normalize=True) * 100

print("Churn counts:\n", churn_counts)
print("\nChurn rate (%):\n", churn_pct.round(2))
print(f"\nOverall churn rate: {(df[TARGET_COLUMN] == 'Yes').mean() * 100:.2f}%")
```

```python
# Missing values & data quality checks
print("Missing values:\n", df.isnull().sum()[df.isnull().sum() > 0])
print(f"\nDuplicate rows: {df.duplicated().sum()}")
print(f"Unique IDs: {df[ID_COLUMN].nunique()} / {len(df)}")
```
